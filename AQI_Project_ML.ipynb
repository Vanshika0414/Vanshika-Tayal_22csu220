```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Load the dataset
file_path = "AirQualityUCI.csv"  # Replace with your file path
data = pd.read_csv(file_path, sep=";", decimal=",", na_values=["NaN", "nan"])

# Combine Date and Time into a single datetime column
try:
    data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d/%m/%Y %H.%M.%S', errors='coerce')
except Exception:
    data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], errors='coerce')

# Filter the data to include only valid rows
data = data[['datetime', 'CO(GT)']].dropna()
data.rename(columns={'datetime': 'ds', 'CO(GT)': 'y'}, inplace=True)
data = data[data['y'] > 0]  # Remove non-positive values

# Ensure at least two valid rows remain
if data.shape[0] < 2:
    raise ValueError("Insufficient valid data for forecasting after preprocessing.")

# Split the data into training and testing sets
train_data = data.iloc[:int(0.8 * len(data))]
test_data = data.iloc[int(0.8 * len(data)):]

# Initialize and fit the Prophet model
model = Prophet()
model.fit(train_data)

# Make future predictions
future = model.make_future_dataframe(periods=len(test_data), freq='H')
forecast = model.predict(future)

# Evaluate the model
y_true = test_data['y'].values
y_pred = forecast['yhat'][-len(test_data):].values

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Plot the forecast
model.plot(forecast)
plt.title("Prophet Forecast")
plt.show()

# Plot the components
model.plot_components(forecast)
plt.show()
```
